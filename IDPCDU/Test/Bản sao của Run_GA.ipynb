```python
import numpy as np
import os
import random
from build import GraphDomain
from GA import *
```

```python
sc = "/content/drive/MyDrive/IDPCDU/Datasets/IDPCDU_Edges/set1/"
TaskNames = os.listdir(sc)
TaskNames.sort()
TaskNames
```

```python
def Load(index):
    path = sc + TaskNames[index]
    name = TaskNames[index].split('.')[0]
    task = GraphDomain(path=path, name=name)
    return task
```

```python
import json

for i in range(16,18):
    history = dict()

    task = Load(i)

    for j in range(30):
        np.random.seed(j)
        random.seed(j)
        task.best = np.inf
        
        ga = GA(100, 500, 0.7, 0.1, task=task)
        log = ga.run()
        history[j] = log

    path = '/content/drive/MyDrive/IDPCDU/Result/GA_result/set1/' + TaskNames[i] + '.json'
    with open(path, "w", encoding="utf-8") as outfile:
        json.dump(history, outfile)
```

```python
import pandas as pd
import json

res = {}

for i in range(24):
    path = '/content/drive/MyDrive/IDPCDU/Result/GA_result/set1/' + TaskNames[i] + ".json"
    log = {}
    with open(path, "r", encoding="utf-8") as infile:
        log = json.load(infile)
    df = pd.DataFrame.from_dict(log)
    df.to_csv (r'/content/drive/MyDrive/IDPCDU/Result/GA_result/set1/'+TaskNames[i]+'.csv', index_label='generation', header=True)
    
    avg = 0
    best = np.inf
    for j in range(30):
        avg += log[str(j)]["499"]
        best = min(best, log[str(j)]["499"])

    res[i] = {"Datasets": TaskNames[i], "BEST": best, "AVG": avg/30}


df = pd.DataFrame.from_dict(res, orient='index')
df.to_csv (r'/content/drive/MyDrive/IDPCDU/Result/GA_result/set1/', index_label='ID', header=True)
```
